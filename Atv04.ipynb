# Exercício 01
<p style="text-align:justify;">
Crie um programa que armazene um vetor com as notas de dez alunos, calcule e imprima a média dessas notas e depois informe a quantidade de notas acima e abaixo da média calculada.
</p>

```java
import java.util.Scanner;

//public class NotasAlunos {

   // public static void main(String[] args) {
        Scanner entrada = new Scanner(System.in);

        // Declaração do vetor de notas
        double[] notas = new double[10];

        // Leitura das notas
        for (int i = 0; i < notas.length; i++) {
            System.out.print("Digite a nota do aluno " + (i + 1) + ": ");
            notas[i] = entrada.nextDouble();
        }

        // Cálculo da média
        double soma = 0;
        for (double nota : notas) {
            soma += nota;
        }
        double media = soma / notas.length;

        // Contagem de notas acima e abaixo da média
        int acimaMedia = 0;
        int abaixoMedia = 0;
        for (double nota : notas) {
            if (nota > media) {
                acimaMedia++;
            } else if (nota < media) {
                abaixoMedia++;
            }
        }

        // Impressão dos resultados
        System.out.println("\nMédia das notas: " + media);
        System.out.println("Notas acima da média: " + acimaMedia);
        System.out.println("Notas abaixo da média: " + abaixoMedia);

        entrada.close();
   // }
//}
```

# Exercício 02
<p style="text-align:justify;">
Construa uma classe em Java que seja capaz de efetuar a leitura de uma matriz de valores reais de tamanho 3 x 3, calculando seu determinante.
</p>

```java
import java.util.Scanner;

//public class MatrizDeterminante {

   // public static void main(String[] args) {
        Scanner entrada = new Scanner(System.in);

        // Declaração e leitura da matriz
        double[][] matriz = new double[3][3];
        for (int i = 0; i < matriz.length; i++) {
            for (int j = 0; j < matriz[i].length; j++) {
                System.out.print("Digite o valor da linha " + (i + 1) + ", coluna " + (j + 1) + ": ");
                matriz[i][j] = entrada.nextDouble();
            }
        }

        // Cálculo do determinante
        double determinante = calcularDeterminante(matriz);

        // Impressão do resultado
        System.out.println("\nDeterminante da matriz:");
        for (int i = 0; i < matriz.length; i++) {
            for (int j = 0; j < matriz[i].length; j++) {
                System.out.print(matriz[i][j] + " ");
            }
            System.out.println();
        }
        System.out.println("\nDeterminante: " + determinante);

        entrada.close();
    }

    private static double calcularDeterminante(double[][] matriz) {
        if (matriz.length != matriz[0].length) {
            throw new IllegalArgumentException("Matriz não é quadrada!");
        }

        if (matriz.length == 1) {
            return matriz[0][0];
        }

        double determinante = 0;
        for (int i = 0; i < matriz[0].length; i++) {
            double cofator = calcularCofator(matriz, 0, i);
            determinante += matriz[0][i] * cofator;
        }

        return determinante;
    }

    private static double calcularCofator(double[][] matriz, int linha, int coluna) {
        if (matriz.length <= linha || matriz[0].length <= coluna) {
            return 0;
        }

        double subMatriz[][] = new double[matriz.length - 1][matriz[0].length - 1];
        for (int i = 0; i < matriz.length; i++) {
            for (int j = 0; j < matriz[0].length; j++) {
                if (i != linha && j != coluna) {
                    int subI = i < linha ? i : i - 1;
                    int subJ = j < coluna ? j : j - 1;
                    subMatriz[subI][subJ] = matriz[i][j];
                }
            }
        }

        return calcularDeterminante(subMatriz) * (matriz[linha][coluna] % 2 == 0 ? 1 : -1);
  //  }
//}
```

# Exercício 04
<p style="text-align:justify;">
Usando um array unidimensional de 10 elementos do tipo String, faça uma classe que simule a ocupação de um estacionamento de veículos. Considere que a posição do array corresponde ao número da vaga e para cada vaga é armazenada a placa do veículo que está estacionado. Sua classe deve conter um laço de repetição simulando um menu que controle a entrada e a saída dos veículos com as opções: 1-Entrada, 2-Saída e 3-Listar situação atual e 4-Encerrar o programa. Se o usuário escolher 1, solicite o número da vaga e a placa do veículo; se escolher 2, solicite o número da vaga que será liberada e armazene o valor branco na posição correspondente ao array; se escolher 3, liste a situação atual apresentando em tela todos os elementos do array; quando for 4, encerre o laço e o programa.
</p>

```java
import java.util.Scanner;

public class Estacionamento {

    private static final int TAMANHO_ESTACIONAMENTO = 10; // Tamanho do array de vagas
    private static String[] vagas = new String[TAMANHO_ESTACIONAMENTO]; // Array de placas de veículos

    public static void main(String[] args) {
        Scanner entrada = new Scanner(System.in);

        // Inicializando vagas como vazias (armazena null)
        for (int i = 0; i < vagas.length; i++) {
            vagas[i] = null;
        }

        int opcao;

        do {
            System.out.println("\n\nMenu do Estacionamento:");
            System.out.println("1 - Entrada de Veículo");
            System.out.println("2 - Saída de Veículo");
            System.out.println("3 - Listar Situação Atual");
            System.out.println("4 - Encerrar Programa");
            System.out.print("Digite sua opção: ");

            opcao = entrada.nextInt();
            entrada.nextLine(); // Consumir quebra de linha

            switch (opcao) {
                case 1:
                    registrarEntradaVeiculo(entrada);
                    break;
                case 2:
                    registrarSaidaVeiculo(entrada);
                    break;
                case 3:
                    listarSituacaoAtual();
                    break;
                case 4:
                    System.out.println("Programa encerrado!");
                    break;
                default:
                    System.out.println("Opção inválida! Tente novamente.");
            }
        } while (opcao != 4);

        entrada.close();
    }

    private static void registrarEntradaVeiculo(Scanner entrada) {
        System.out.print("Digite o número da vaga: ");
        int vaga = entrada.nextInt();
        entrada.nextLine(); // Consumir quebra de linha

        if (vaga < 1 || vaga > TAMANHO_ESTACIONAMENTO) {
            System.out.println("Vaga inválida! Tente novamente.");
            return;
        }

        if (vagas[vaga - 1] != null) {
            System.out.println("Vaga ocupada! Tente outra vaga.");
            return;
        }

        System.out.print("Digite a placa do veículo: ");
        String placa = entrada.nextLine();

        vagas[vaga - 1] = placa;
        System.out.println("Veículo estacionado com sucesso na vaga " + vaga + "!");
    }

    private static void registrarSaidaVeiculo(Scanner entrada) {
        System.out.print("Digite o número da vaga: ");
        int vaga = entrada.nextInt();
        entrada.nextLine(); // Consumir quebra de linha

        if (vaga < 1 || vaga > TAMANHO_ESTACIONAMENTO) {
            System.out.println("Vaga inválida! Tente novamente.");
            return;
        }

        if (vagas[vaga - 1] == null) {
            System.out.println("Vaga vazia! Nenhum veículo estacionado.");
            return;
        }

        vagas[vaga - 1] = null;
        System.out.println("Veículo removido da vaga " + vaga + " com sucesso!");
    }

    private static void listarSituacaoAtual() {
        System.out.println("\nSituação atual do estacionamento:");
        for (int i = 0; i < vagas.length; i++) {
            System.out.println("Vaga " + (i + 1) + ": " + (vagas[i] != null ? vagas[i] : "Vaga vazia"));
        }
    }
}
```

# Exercício 05
<p style="text-align:justify;">
Uma imagem é formada por pixels. Considere uma imagem com dimensão de 40 x 40 e faça uma classe que contenha um array bidimensional com essas dimensões. A seguir, para cada posição
desse array bidimensional armazene um valor aleatório entre 0 e 255 (esses valores correspondem às tonalidades aplicadas sobre a imagem). Apresente em tela os 1600 valores gerados.
</p>

```java
import java.util.Random;

public class ImagemPixels {

    private static final int LARGURA = 40; // Largura da imagem
    private static final int ALTURA = 40; // Altura da imagem
    private static final int TAMANHO_MAXIMO_COR = 255; // Valor máximo para a tonalidade de cada pixel

    private static int[][] imagem; // Array bidimensional para armazenar os valores de cada pixel

    public static void main(String[] args) {
        Random random = new Random();

        // Criando e inicializando o array bidimensional
        imagem = new int[ALTURA][LARGURA];
        for (int i = 0; i < ALTURA; i++) {
            for (int j = 0; j < LARGURA; j++) {
                imagem[i][j] = random.nextInt(TAMANHO_MAXIMO_COR + 1); // Gera valor aleatório entre 0 e 255
            }
        }

        // Imprimindo os valores dos pixels
        for (int i = 0; i < ALTURA; i++) {
            for (int j = 0; j < LARGURA; j++) {
                System.out.print(imagem[i][j] + " ");
            }
            System.out.println();
        }
    }
}
```
